```python
import tensorflow
from keras.applications.resnet import ResNet50, preprocess_input
from keras.layers import GlobalMaxPooling2D
from tensorflow.keras.preprocessing import image
```

```python
model=ResNet50(weights='imagenet',include_top=False,input_shape=(224,224,3))
model.trainable=False
```

```python
model=tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
model.summary()
```

```python
import cv2
import numpy as np
from numpy.linalg import norm
```

```python
img=cv2.imread("C:/Users/91866/OneDrive/Desktop/sweater.jpg")
img=cv2.resize(img,(224,224))
```

```python
def extract_features(img_path,model):
    img = image.load_img(img_path,target_size=(224,224))
    img_array = image.img_to_array(img)
    expanded_img_array = np.expand_dims(img_array, axis=0)
    preprocessed_img = preprocess_input(expanded_img_array)
    result = model.predict(preprocessed_img).flatten()
    normalized_result = result / norm(result)

    return normalized_result
    
    
```

```python
import os
from tqdm import tqdm
```

```python
filenames=[]
feature_list=[]

for file in os.listdir("C:/Users/91866/OneDrive/Desktop/visuelle_reccomender"):
    filenames.append(os.path.join("C:/Users/91866/OneDrive/Desktop/visuelle_reccomender",file))
```

```python
filenames[0:5]
```

```python
for file in tqdm(filenames):
    feature_list.append(extract_features(file,model))
    
```

```python
import pickle

pickle.dump(feature_list,open("feature_vector.pkl","wb"))
pickle.dump(filenames,open('file_names.pkl','wb'))

```
